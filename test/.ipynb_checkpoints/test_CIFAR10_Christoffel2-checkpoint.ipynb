```julia
using CSV
using DataFrames

df = CSV.read("/home/hoanganh/Desktop/math-topics/algebraic_statistics/codes/datasets/train_CIFAR10.csv", DataFrame)
```

```julia
nr=50000
nc=3073
D=Matrix{Float64}(undef,nr,nc)
for i=1:nr
    for j=1:nc
        D[i,j]=df[i,j]
    end
    D[i,nc]+=1
end
```

```julia
max_col=[maximum(D[:,j]) for j=1:nc-1]
```

```julia
for j=1:nc-1
    D[:,j]/=max_col[j]
end
D
```

```julia
D[:,1:nc-1].-=0.5
D[:,1:nc-1]*=2
```

```julia
using LinearAlgebra

max_norm_col=maximum(norm(D[j,1:nc-1]) for j=1:nr) 
```

```julia
r=1
D[:,1:nc-1]/=max_norm_col/r
D
```

```julia
Y=Vector{Matrix{Float64}}(undef,10)

for k in 1:10
    Y[k]=D[findall(u -> u == k, D[:,end]),1:nc-1]
end
N=nc-1
```

```julia
t=Vector{Int64}(undef,10)
Y_train=Vector{Matrix{Float64}}(undef,10)

for k=1:10
    t[k]=ceil(Int64,0.9*size(Y[k],1))
    Y_train[k]=Y[k][1:t[k],:]
end

Y_train[1]
```

```julia
Y_test=Vector{Matrix{Float64}}(undef,10)

for k=1:10
    Y_test[k]=Y[k][(t[k]+1):end,:]
end
```

```julia
using SparseArrays

eps=1e-2

invM=Vector{Matrix{Float64}}(undef,10)
u=zeros(Float64,N)
v=zeros(Float64,2*N+1)
for k=1:10
    invM[k]=zeros(Float64,2*N+1,2*N+1)
    for j in 1:t[k]
        u=Y_train[k][j,:]
        v=[1;u;u.^2]
        invM[k]+=v*v'
    end
    invM[k]=inv(invM[k]/t[k]+eps*sparse(I, 2*N+1, 2*N+1)) 
end
```

```julia
function create_CF(invM) 
    function CF(y)       
        eval_momo=[1;y;y.^2]
        val=eval_momo'*invM*eval_momo
        return 1/val
    end
    return CF
end
```

```julia
Lambda=Vector{Function}(undef,10)

for k=1:10
    println("Class ",k)
    println()
    
    Lambda[k]=create_CF(invM[k])
end
```

```julia
function classifier(y)
    return findmax([Lambda[k](y) for k=1:10])[2]
end
```

```julia
predict=Vector{Vector{Int64}}(undef,10)

for k=1:10
    predict[k]=[classifier(Y_test[k][j,:]) for j in 1:size(Y_test[k],1)]
end
```

```julia
numcor=Vector{Int64}(undef,10)

for k=1:10
    numcor[k]=length(findall(u -> u == k, predict[k]))
end
```

```julia
sum(numcor)
```

```julia
accuracy=(sum(numcor))/(sum(size(Y_test[k],1) for k=1:10))
```
