```julia
#=
N=2
t=100
Y=Matrix{Float64}(undef,t,N)
randx=zeros(Float64,N)
for j=1:t
    randx=2*rand(Float64,N).-1
    randx=rand(Float64)*randx./sqrt(sum(randx.^2))
    Y[j,:]=randx
end
=#
```

```julia
#=
data="/home/hoanganh/Desktop/math-topics/algebraic_statistics/codes/datasets/plot" # path of data
output_file = open(data*"/bivariate_MLE_density_data.jl","w")

write(output_file, "N = $(N)")
write(output_file, "; \n \n")
write(output_file, "t = $(t)")
write(output_file, "; \n \n")
write(output_file, "Y = ")
show(output_file, Y)
write(output_file, "; \n \n")

close(output_file)
=#
```

```julia
include("../src/ClassificationModels.jl")
using .ClassificationModels

R=1

x=ClassificationModels.solve_opt(N,Y,t,R,d,delta=1,s=1,rho=1,numiter=1e5,eps=1e-4,tol_eig=1e-3,ball_cons=true);
```

```julia
include("../src/ClassificationModels.jl")
using .ClassificationModels

eval_PDF=ClassificationModels.func_eval_PDF(x,N,d,R,ball_cons=true)
```

```julia
using Plots

import Contour: contours, levels, level, lines, coordinates

p_approx(x1,x2)=eval_PDF([x1;x2])

x1s = -1:0.01:1
x2s = -1:0.01:1

contour(x1s, x2s, p_approx,fill=true,title = "d = $(d)")

scatter!(Y[:,1], Y[:,2],label ="sample")
```
