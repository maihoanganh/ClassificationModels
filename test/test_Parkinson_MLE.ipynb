```julia
using CSV
using DataFrames

df = CSV.read("/home/hoanganh/Desktop/math-topics/algebraic_statistics/codes/datasets/ReplicatedAcousticFeatures-ParkinsonDatabase.csv", DataFrame)
```

```julia
nr=240
nc=46
D=Matrix{Float64}(undef,nr,nc)
for j=1:nr
    for i=4:48
        D[j,i-3]=df[j,i]
    end
        
    D[j,46]=df[j,3]

end
D
```

```julia
max_col=[maximum(D[:,j]) for j=1:45]
```

```julia
for j=1:45
    D[:,j]/=max_col[j]
end
D
```

```julia
D[:,1:45].-=0.5
D[:,1:45]*=2
```

```julia
using LinearAlgebra

max_norm_col=maximum(norm(D[j,1:45]) for j=1:nr) 
```

```julia
r=1.7
D[:,1:45]/=max_norm_col/r
D
```

```julia
Y1=D[findall(u -> u == 0, D[:,end]),1:45]
N=45
Y1
```

```julia
Y2=D[findall(u -> u == 1, D[:,end]),1:45]
```

```julia
t1=ceil(Int64,0.9*size(Y1,1))
Y_train1=Y1[1:t1,:]
```

```julia
t2=ceil(Int64,0.9*size(Y2,1))
Y_train2=Y2[1:t2,:]
```

```julia
Y_test1=Y1[(t1+1):end,:]
```

```julia
Y_test2=Y2[(t2+1):end,:]
```

```julia
include("../src/ClassificationModels.jl")
using .ClassificationModels

println("Class 1")
println()

d1=1
R1=r

x1=ClassificationModels.solve_opt(N,Y_train1,t1,R1,d1;delta=1,s=1,rho=1,
                         numiter=1e4,eps=-1e-3,tol_eig=1e-3,ball_cons=true,feas_start=false);
println()
println("Class 2")
println()
d2=1
R2=r

x2=ClassificationModels.solve_opt(N,Y_train2,t2,R2,d2;delta=1,s=1,rho=1,
                         numiter=1e4,eps=-1e-3,tol_eig=1e-3,ball_cons=true,feas_start=false);
```

```julia
include("../src/ClassificationModels.jl")
using .ClassificationModels

eval_PDF1=ClassificationModels.func_eval_PDF(x1,N,d1,R1,ball_cons=true)
```

```julia
include("../src/ClassificationModels.jl")
using .ClassificationModels

eval_PDF2=ClassificationModels.func_eval_PDF(x2,N,d2,R2,ball_cons=true)
```

```julia
function classifier(y)
    if eval_PDF1(y)>eval_PDF2(y)
        return 1
    else
        return 2
    end
end
```

```julia
predict1=[classifier(Y_test1[j,:]) for j in 1:size(Y_test1,1)]
```

```julia
numcor1=length(findall(u -> u == 1, predict1))
```

```julia
predict2=[classifier(Y_test2[j,:]) for j in 1:size(Y_test2,1)]
```

```julia
numcor2=length(findall(u -> u == 2, predict2))
```
